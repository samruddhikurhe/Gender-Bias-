```python
import pandas as pd
import random
```

```python
# Load the male similarity Excel file
male_similarity_df = pd.read_excel("male_similarity.xlsx")

# Preview the dataframe
male_similarity_df.head()
```

```python
# Ensure enough data
assert len(male_similarity_df) >= 50, "Not enough pairs in the file!"

# Randomly sample 50 pairs
male_50_pairs = male_similarity_df.sample(n=50, random_state=42).reset_index(drop=True)

# Show sample
male_50_pairs.head()

```

```python
# Save to Excel
male_50_pairs.to_excel("male_50_random_pairs.xlsx", index=False)
print("ðŸ’¾ File saved as 'male_50_random_pairs.xlsx'")

```
